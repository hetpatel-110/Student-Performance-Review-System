```python
def new_password():
    security.destroy()
    global new_password_screen
    new_password_screen=Toplevel(main_screen)
    new_password_screen.title("New Password")
    new_password_screen.geometry("400x400")

    Label(new_password_screen, text="Enter New Password : ", font=("Consolas", 12, "bold")).grid(row=0, column=0)
    
    global new_password_entry
    new_password_entry=Entry(new_password_screen, width=30, font=('Consolas', 12))
    new_password_entry.grid(row=0, column=1, padx=10, pady=10)

    Label(new_password_screen, text="Confirm Password : ", font=("Consolas", 12, "bold")).grid(row=1, column=0, pady=10)

    global new_password_confirm_entry
    new_password_confirm_entry=Entry(new_password_screen, width=30, font=('Consolas', 12))
    new_password_confirm_entry.grid(row=1, column=1, padx=10, pady=10)

    global confirmpass
    global newpass
    confirmpass = new_password_confirm_entry.get()
    newpass = new_password_entry.get()
    
    def check():
        if new_password_confirm_entry.get()==new_password_entry.get():
            new_password_submit()
            new_password_screen.destroy()
            security.destroy()
        else:
            Label(new_password_screen, text="Passwords do not Match!", font=("Consolas", 11, "bold")).grid(row=2, column=0, pady=10)

    Button(new_password_screen, text="SAVE PASSWORD", command=check).grid(row=2, column=0, columnspan=2)
    
def new_password_submit():
    global new_password_commit
    new_password_commit=Toplevel(main_screen)
    new_password_commit.title("Change Password")
    new_password_commit.geometry("400x400")
    conn=sqlite3.connect("project.db")
    c=conn.cursor()

    if(category=="STUDENT"):
        c.execute('UPDATE STUDENT SET PASSWORD="%s" WHERE ROLLNUMBER="%s"'%(new_password_entry.get(),sec_id))
    else:
        c.execute('UPDATE ADMIN SET PASSWORD="%s" WHERE USERNAME="%s"'%(new_password_entry.get(),sec_id))

    conn.commit()
    c.close()
    Label(new_password_commit, text="Password Changed Successfully!!", font=("Consolas", 11, "bold")).pack()
    Button(new_password_commit, text="OK", command=new_password_commit.destroy).pack()
```

SECURITY QUESTION SCREEN

```python
def security_question():
    password_not_recog_screen.destroy()

    global security
    security=Toplevel(main_screen)
    security.title("Security Question")
    security.geometry("700x500")

    Label(security, text="VERIFY AS A VALID USER", font=("Consolas", 13, "bold")).pack()
    Label(security, text="").pack()

    conn=sqlite3.connect("project.db")
    c=conn.cursor()
    c.execute('SELECT QUESTION FROM SECURITY_QUESTION')
    que = c.fetchall()
    c.close()
    
    ques=[list(i) for i in que]
    ques=[i[0] for i in ques]
    
    categ = ttk.Combobox(security, state="readonly", width=17, value=["STUDENT", "ADMIN"], font=("Consolas", 13, "bold"))
    def_categ="SELECT CATEGORY"

    categ.set(def_categ)
    categ.pack(pady=30)

    drop = ttk.Combobox(security, state="readonly", width=37, value=ques, font=("Consolas", 13, "bold"))
    def_drop="SELECT YOUR SECURITY QUESTION"
    
    drop.set(def_drop)
    drop.pack()

    global answer
    answer=StringVar()
    answer=Entry(security, width=30, font=('Consolas', 12,'bold'))
    answer.pack(pady=30)

    def getopt():
        global category
        global sec_que
        global sec_ans
        category=categ.get()
        sec_que=drop.get()
        sec_ans=answer.get()

        if category=="STUDENT":
            student_sec()
        else:
            admin_sec()

    def student_sec():
        conn=sqlite3.connect("project.db")
        c=conn.cursor()
        c.execute('SELECT SECURITY_Q, SECURITY_A FROM STUDENT WHERE ROLLNUMBER="%s"'%(sec_id))

        rec = c.fetchall()
        if sec_que==rec[0][0]:
            if sec_ans==rec[0][1]:
                new_password()
            else:
                invalid()
        else:
            invalid()

        c.close()

    def admin_sec():
        conn=sqlite3.connect("project.db")
        c=conn.cursor()
        c.execute('SELECT SECURITY_Q, SECURITY_A FROM ADMIN WHERE USERNAME="%s"'%(sec_id))

        rec = c.fetchall()
        print(rec)
        if sec_que==rec[0][0]:
            if sec_ans==rec[0][1]:
                new_password()
            else:
                invalid()
        else:
            invalid()

        c.close()

    but=Button(security, text="VERIFY", command=getopt)
    but.pack(pady=30)
```

STUDENT LOGIN PAGE

```python
# Designing window for login 

def stu_login():
    global stu_login_screen
    stu_login_screen = Toplevel(main_screen)
    stu_login_screen.title("Student Login")
    stu_login_screen.geometry("700x500")

    pic1=Image.open("stu.png")
    height1=200
    width1=int(height1/pic1.height * pic1.width)
    resized1=pic1.resize((width1,height1),Image.ANTIALIAS)
    global img1
    img1=ImageTk.PhotoImage(resized1)
    Label(stu_login_screen, image=img1).pack()

    Label(stu_login_screen, text="Enter Roll Number and Password").pack()
    Label(stu_login_screen, text="").pack()
 
    global rollnumber_verify
    global password_verify

    rollnumber_verify = StringVar()
    password_verify = StringVar()

    global rollnumber_stu_login_entry
    global password_stu_login_entry
 
    Label(stu_login_screen, text="Roll Number", font="Consolas 15 bold").pack()
    rollnumber_stu_login_entry = Entry(stu_login_screen, textvariable=rollnumber_verify, font="Consolas 15 bold")
    rollnumber_stu_login_entry.pack()

    Label(stu_login_screen, text="").pack()
    
    Label(stu_login_screen, text="Password", font="Consolas 15 bold").pack()
    password_stu_login_entry = Entry(stu_login_screen, textvariable=password_verify, show='*', font="Consolas 15 bold")
    password_stu_login_entry.pack()

    Label(stu_login_screen, text="").pack()

    Button(stu_login_screen, text="Login", height=1, command = stu_login_verify, font="Consolas 13 bold", borderwidth=2, relief=SOLID, background="skyblue").pack()
    Label(stu_login_screen, text="").pack()
    
# Implementing event on stu_login button 
 
def stu_login_verify():
    global rollnumber
    rollnumber = rollnumber_verify.get()
    password = password_verify.get()
    rollnumber_stu_login_entry.delete(0, END)
    password_stu_login_entry.delete(0, END)

    if rollnumber=='' or password=='':
        print("Fill the empty field!!!")
    else:
        #open database
        conn = sqlite3.connect('project.db')
        #select query
        c = conn.execute('SELECT * from STUDENT where ROLLNUMBER="%s" and PASSWORD="%s"'%(rollnumber,password))
        #fetch data 
        if c.fetchone():
           stu_login_sucess()
        else:
            c = conn.execute('SELECT * from STUDENT where ROLLNUMBER="%s"'%(rollnumber))
            if c.fetchone():
                global sec_id
                sec_id=rollnumber
                password_not_recognised()
            else:
                user_not_found()
        
# Designing popup for stu_login success
 
def stu_login_sucess():
    global stu_login_success_screen
    stu_login_success_screen = Toplevel(stu_login_screen)
    stu_login_success_screen.title("Success")
    stu_login_success_screen.geometry("150x100")
    Label(stu_login_success_screen, text="Login Success").pack()
    Button(stu_login_success_screen, text="PROCEED", command=stu_login_success_del).pack()
```
